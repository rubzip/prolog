# Converting our sk file

```python
import re
pattern = r"\w+\((\d+)\,\d+\,\'(\w+)\%\d+\:(\d+)"

sks = []
with open("wn_sk.pl", 'r') as f:
    for line in f:
        match = re.search(pattern=pattern, string=line)
        if match:
            sks.append(f"sk({match.group(1)}, '{match.group(2)}', {match.group(3)}).\n")
```

```python
with open("wn_sk_ruben.pl", 'w') as f:
    for sk in sks:
        f.write(sk)
```

# Adding Semantic Rules

```python
dictionary = pd.read_csv("./dictionary.csv")
translator = dict(zip(dictionary["Name"], dictionary["File Number"]))
```

```python
df = pd.read_csv("subject_verb_object.csv")
df
```

```python
aux = df.copy()
for col in aux.columns:
    aux[col] = aux[col].map(translator)
aux
```

```python
with open("svo.pl", 'w') as f:
    for line in aux.values:
        f.write(f"s_v_o({line[0]}, {line[1]}, {line[2]}).\n")
```
